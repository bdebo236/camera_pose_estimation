```python
import pandas as pd
import cv2
import os
```

```python
video_path = "../data/videos/IMG_0171.mp4"
output_dir = "../data/frames/IMG_0171_frames"
os.makedirs(output_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
idx = 0
while True:
    ret, frame = cap.read()
    if not ret: break
    cv2.imwrite(f"{output_dir}/frame_{idx:04d}.png", frame)
    idx += 1
cap.release()
print(f"Extracted {idx} frames")
```
